```python
#Find the chord distance
import itertools

def chorddist(chord1,chord2,distmat):
    notenames = ['A','A#','B','C','C#','D','D#','E','F','F#','G','G#'];
    P = list(itertools.permutations(chord2))
    minD = 10000;
    for ii in range(len(P)):  

        distance = 0;
        for i in range(len(chord1)):
            #print(notenames.index(P[ii][i]))
            #print(notenames.index(chord1[i]))
            distance = distance + distmat[notenames.index(chord1[i])][notenames.index(P[ii][i])];
        if distance < minD:
            minD = distance;
    dist = minD
    return dist
```

```python
def getTransMatrix(set1,set2,distmat):
    transMat = np.zeros([len(set1),len(set2)]);
    for i in range(len(set1)):
        for j in range(len(set2)):
            chord1 = printChord(set1[i]);
            chord2 = printChord(set2[j]);
            transMat[i,j] = chorddist(chord1,chord2,distmat);
    return transMat
```
